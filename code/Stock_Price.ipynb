```python
def stock_price(ticker):
    try:
        if not ticker:
            raise ValueError("Ticker symbol must be provided")
        
        url = f'https://finance.naver.com/item/main.naver?code={ticker}'
        res = requests.get(url)
        
        # HTTP 요청 상태 코드 확인
        if res.status_code != 200:
            raise Exception(f"Failed to retrieve data: {res.status_code}")
        
        soup = bts(res.text, 'html.parser')
        items = soup.select('p.no_today > em > span')
        
        if not items:
            raise Exception("No stock price information found in Naver Stock Service")
        
        return items[0].text.strip()
    
    except requests.RequestException as e:
        return f"Request error: {e}"
    except Exception as e:
        return f"An error occurred: {e}"
```

```python
# 회사 이름과 티커 딕셔너리
companies = {
    '삼성전자': '005930',
    '하이닉스': '000660',
    '엘지엔솔': '373220',
    '현대차': '005380',
    '삼성바이오로직스': '207940'
}

for compNm, ticker in companies.items():
    price = stock_price(ticker)
    print(f'{compNm}의 현재 주가는 {price}원 입니다.')
```
